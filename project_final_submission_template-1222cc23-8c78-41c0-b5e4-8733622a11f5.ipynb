### Step 2a: Building
#### Document which information you will represent in your data definitions

The most important information that my program will be reading is country name, Cropland Footprint, Grazing Footprint, Forest Footprint, and Total Ecological Footprint. The country name will tell the program from which row of data we are reading. The Cropland, Grazing, and Forest Footprints will be added together for each selected country to represent the total "Land Footprint". I need the Total Ecological Footprint in order to determine what percentage the newly created Land Footprint makes up of the Total Ecological Footprint. 

#### Design data definitions

I need two data definitions. One for storing that country's cropland/grazing/forest/total ecological footprints/name and one for interpreting the list.


```python
from cs103 import *
from typing import NamedTuple, List
import csv

##################
# Data Definitions
                                           
Country_Footprint = NamedTuple ('Country_Footprint', [('cropland', float),
                                  ('grazing', float),
                                  ('forest', float),
                                  ('total_eco', float),
                                  ('country_name', str)])

# interp. a country's cropland footprint, grazing footprint, and 
# forest footprint, total ecological footprint in gha as well as the country's name. 

CF1 = (0.82, 0.27, 0.63, 6.16, 'Algeria')
CF2 = (0.49, 0.04, 0.26, 4.01, 'Bahamas')

@typecheck
def fn_for_country_footprint(cf: Country_Footprint)->...: #template based on Compound
    return ...(cf.cropland, 
               cf.grazing,
               cf.forest,
               cf.total_eco,
               cf.country_name)

# List[Country_Footprint]
# intrep. a list of country's footprints

L0 = []
L1 = [CF1, CF2]

#template based on arbitrary-sized and reference rule
@typecheck
def fn_for_locf(locf: List[Country_Footprint]) -> ...:
    # description of the acc
    acc = ...  # type: ...
    for cf in locf:
        acc = ...(fn_for_country_footprint(cf), acc)   
    return ...(acc)
```

### Step 2b and 2c: Building
#### Design a function to read the information and store it as data in your program
#### Design functions to analyze the data


Complete these steps in the code cell below. You will likely want to rename the analyze function so that the function name describes what your analysis function does.

```python
import csv
import matplotlib.pyplot as plt
from typing import List
###########
# Functions

@typecheck
def main(filename: str) -> None:
    """
    reads the information from the given filename and return a pie
    chart showing the percentage that land footprint makes up of the total ecological footprint 
    for the given countries.
    """
    #return None #stub
    # Template from HtDAP based on a function composition
    return multiple_charts(read(filename))

@typecheck
def read(filename: str) -> List[Country_Footprint]:
    """    
    reads information from the specified file and returns a list of footprints and country names.
    """
    #return []  #stub
    # Template from HtDAP
    # locf contains the result so far
    locf = [] # type: List[Country_Footprint]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            cf = Country_Footprint(parse_float(row[5]), parse_float(row[6]), parse_float(row[7]), parse_float(row[10]), row[0])
            locf.append(cf)
    
    return locf
    
@typecheck
def show_pie_chart(f: float, country_name:str) -> None:
    """
    show a pie chart with the given labels and corresponding percentages of how each country's 
    land footprint makes up it's total ecological footprint.
    percentages must sum to 100 and so must have at least one element. returns None. 
    """
    #return None  #stub
    # Template based on visualization
    
    data = [f, 100.0-f] 
    fig1, ax1 = plt.subplots()
    ax1.pie(data, autopct ='%1.1f%%', labels= ['Land Footprint', 'Leftover Footprint'])
    ax1.axis('equal') 
    plt.title(country_name)
    
    plt.show()
    
    return None

@typecheck
def multiple_charts(locf: List[Country_Footprint]) -> None:
    """
    creates a chart for each country in list. 
    """
    for cf in locf:
        if has_data(cf):
            show_pie_chart(land_footprint(cf), cf.country_name) 

@typecheck
def has_data(cf: Country_Footprint) -> bool:
    """
    returns False if any of the data for the given country is blank, otherwise True.
    """
    try: 
        (cf.cropland + cf.grazing + cf.forest)/(cf.total_eco)
    except: 
        return False 
    return True
    
@typecheck
def only_country_names(locf:List[Country_Footprint]) -> List[str]:
    """
    returns a list of strings from locf for each country name.
    """
    #return [] #stub
    # template from List[Country_Footprint]
    # acc contains the result so far
    acc = [] # type: List[Country_Footprint]
    for cf in locf:
        acc.append(cf.country_name)
    return acc
    
@typecheck
def only_land_footprint(locf:List[Country_Footprint]) -> List[float]:
    """
    returns a list of floats from locf for given country.
    """
    #return [] #stub
    # template from List[Country_Footprint]
    # acc contains the result so far
    acc = [] # type: List[Country_Footprint]
    for cf in locf:
        acc.append(land_footprint(cf))
    return acc

@typecheck
def land_footprint(cf: Country_Footprint) -> float:
    """
    adds the cropland, grazing, and forest footprint values together for a given country to create a "land footprint" value
    divided by the total ecological footprint value mulitplied by 100 and rounded to to give
    percentage that the land footprint makes up of total ecological one. 
    """
    #return 10.0 #stub
    # template from Country_Footprint
    
    return round(((cf.cropland + cf.grazing + cf.forest)/(cf.total_eco) * 100))

# Begin testing
start_testing()

# Examples and tests for main
expect(main("global_ecological_footprint_UBC_test1.csv"), None)
expect(main("global_ecological_footprint_UBC_test2.csv"), None)

# Examples and tests for read
expect(read("global_ecological_footprint_UBC_test1.csv"), [Country_Footprint(0.3, 0.2, 0.08, 0.79, 'Afghanistan'), Country_Footprint(0.78,0.22,0.25, 2.21, 'Albania'), Country_Footprint(0.6, 0.16,0.25,2.12,'Algeria')])
expect(read("global_ecological_footprint_UBC_test2.csv"), [Country_Footprint(0.2, 0.32, 0.29, 1.37, 'Zimbabwe'), Country_Footprint(0.19, 0.18, 0.33, 0.99, 'Zambia'), Country_Footprint(0.34, 0.14, 0.04, 1.03, 'Yemen')])

# Examples and tests for show_pie_chart
expect(show_pie_chart(15, 'Algeria'), None)
expect(show_pie_chart(22, 'Zimbabwe'), None)

# Examples and tests for multiple_charts
expect(multiple_charts([Country_Footprint(0.2, 0.32, 0.29, 1.37, 'Zimbabwe'), Country_Footprint(0.19, 0.18, 0.33, 0.99, 'Zambia'), Country_Footprint(0.34, 0.14, 0.04, 1.03, 'Yemen')]), None)
expect(multiple_charts([Country_Footprint(0.3, 0.2, 0.08, 0.79, 'Afghanistan'), Country_Footprint(0.78,0.22,0.25, 2.21, 'Albania'), Country_Footprint(0.6, 0.16, 0.25,2.12,'Algeria')]), None)

# Examples and tests for has_data
expect(has_data(Country_Footprint(0.2, 0.32, 0.29, 1.37, 'Zimbabwe')), True)
expect(has_data(Country_Footprint(0, 0, 0, 0, 'Aruba')), False)

# Examples and tests for only_country_names
expect(only_country_names([Country_Footprint(0.2, 0.32, 0.29, 1.37, 'Zimbabwe')]), ['Zimbabwe'])
expect(only_country_names([Country_Footprint(0.3, 0.2, 0.08, 0.79, 'Afghanistan')]), ['Afghanistan'])

# Examples and tests for only_land_footprint
expect(only_land_footprint ([Country_Footprint(0.2, 0.32, 0.29, 1.37, 'Zimbabwe')]), [59])
expect(only_land_footprint ([Country_Footprint(0.78, 0.22, 0.25, 2.21,'Albania')]), [57])

# Examples and tests for land_footprint
expect(land_footprint(Country_Footprint(0.78, 0.22, 0.25, 2.21,'Albania')), 57)
expect(land_footprint(Country_Footprint(0.2, 0.32, 0.29, 1.37, 'Zimbabwe')), 59)

# show testing summary 
summary()
```
